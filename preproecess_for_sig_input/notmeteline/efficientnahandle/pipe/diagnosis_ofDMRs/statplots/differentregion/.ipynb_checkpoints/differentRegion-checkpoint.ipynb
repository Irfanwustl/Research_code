```python
import pandas as pd
ingenefile="/Users/irffanalahi/Research/Research_update/SM/melcfdref/LargerLTME/BluOurPBL/Dashboard/BlOurPBLourCD8TIL_out_mincpg2_q0.5_addedcol_q0.00001_diff0.1_hypo_noheader_genomic_feature_withrepeat_header_celltypeseperated/develop/CD8TIL_row_combined.txt"
outfile=ingenefile+"_prioritizedregion.txt"
priority_list=['promoter','cds','3utrexon','5utrexon','intron','repeat']

ingenedf=pd.read_csv(ingenefile,sep="\t")

ingenedf.head()
```

```python
grouped=ingenedf.groupby(['chrom','start','end'])
rowlist=[]
for name, group in grouped:
    for region in priority_list:
        tmp= group[group['Region']==region]
        if tmp.shape[0]>0:
            tmprow=tmp.iloc[0]
            if tmprow['Region']=='3utrexon' or tmprow=='5utrexon':
                tmprow['Region']=='3/5 utrexon'
                
            rowlist.append(tmp.iloc[0])  ##### hadle 2utr/5ut
            break
```

```python
outdf=pd.DataFrame(rowlist)
outdf.to_csv(outfile,sep="\t",index=False)
```
