```python
import pandas as pd
ingenefile="/Users/irffanalahi/Research/Research_update/SM/melcfdref/bestworkingref/genomic_analysis/pbltilmel_mincp2outcutoff_addedcol_genomic_feature_withrepeat_header_celltypeseperated/TIL_row_combined.txt"
outfile=ingenefile+"_prioritizedregion.txt"
priority_list=['promoter','cds','3utrexon','5utrexon','intron','repeat']#['repeat','cds','3utrexon','5utrexon','promoter','intron']#['promoter','cds','3utrexon','5utrexon','intron','repeat']

ingenedf=pd.read_csv(ingenefile,sep="\t")

ingenedf.head()
```

```python
grouped=ingenedf.groupby(['chrom','start','end'])
rowlist=[]
for name, group in grouped:
    for region in priority_list:
        tmp= group[group['Region']==region]
        if tmp.shape[0]>0:
            tmprow=(tmp.iloc[0]).copy()
            
            if (tmprow['Region']=='3utrexon') | (tmprow['Region']=='5utrexon'):
                tmprow['Region']='3/5 utrexon'
                
            rowlist.append(tmprow)  
            break
```

```python
outdf=pd.DataFrame(rowlist)
outdf=outdf[['chrom','start','end','Region']]
duplicateRowsDF = outdf[outdf.duplicated()]
if duplicateRowsDF.shape[0]>0:
    print("duplicated rows.WHY???? FIX it before plotting")
    print(duplicateRowsDF)
outdf.to_csv(outfile,sep="\t",index=False)
```
