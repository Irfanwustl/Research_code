```python
import pandas as pd

infile='/Users/irffanalahi/Research/Research_update/troubleshootoffbyoneissue/bedstats/zzmel-167-P-CD8.bedgraph_rolled.bedgraph'


pad=1

indf=pd.read_csv(infile,sep='\t',header=None)
indf.head()
```

```python
indf['lesspaddedstart']=indf[1]-pad
indf['lesspaddedend']=indf[2]-pad
indf['greaterpaddedstart']=indf[1]+pad
indf['greaterpaddedend']=indf[2]+pad
indf.head()
```

```python
outlessadded=indf[[0,'lesspaddedstart','lesspaddedend',3]]
outgreaterpadded=indf[[0,'greaterpaddedstart','greaterpaddedend',3]]
```

```python
outlessadded.to_csv(infile+"_less_"+str(pad),sep='\t',index=False,header=False)
outgreaterpadded.to_csv(infile+"_greater_"+str(pad),sep='\t',index=False,header=False)
```
