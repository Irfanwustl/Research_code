```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from scipy.optimize import fsolve
from scipy.stats import binom
import matplotlib.ticker as mtick
```

```python
groupsize = 3
mincpg = 5

FP_col = 'false_pos/all_neg'
mNeu_col = 'mNeu_real'
```

```python
infile = '92-reporter.txt'
data = pd.read_csv(infile, sep='\t', index_col=0)
df = pd.DataFrame(data)

df = df[df['mincpg'] == mincpg]

df['Index Number'] = [int(i.split('insilmix')[-1]) for i in df.index]

df.sort_values(['Index Number'], inplace=True)
df = df.drop('Index Number', axis=1)

df
```

```python
mNeu_real = []
FP = []

count = 0
mNeu_sub = []
FP_sub = []
for i in range(len(df)):
    mNeu_sub.append(float(df.iloc[i][mNeu_col]))
    FP_sub.append(float(df.iloc[i][FP_col]))
    if count == groupsize:
        mNeu_real.append(np.mean(mNeu_sub))
        FP.append(np.mean(FP_sub))
        mNeu_sub = []
        FP_sub = []
        count = 0
    count += 1
    
zipped_lists = zip(mNeu_real, FP)
sorted_pairs = sorted(zipped_lists)

tuples = zip(*sorted_pairs)
mNeu_real, FP = [ list(tuple) for tuple in  tuples]

fig, ax = plt.subplots(1, 1)
ax.plot(mNeu_real, FP, c='k', alpha=0.5, ls='--')
ax.fill_between(mNeu_real, FP, color='k', alpha=0.3)
ax.invert_xaxis()
ax.set_xscale('log')
```

```python
adj = 0.95
out = 'multi_reporters.pdf'

f1 = np.array(mNeu_real) / 100
r1 = [(16, 'A'), (23, 'B'), (29, 'C'), (31, 'D'), (47, 'E'), (100, 'F'), (110, 'G'), (216, 'H'), (340, 'I')]
min_reporters = [mincpg]
d1 = 30
```

```python
fig, axs = plt.subplots(1, len(min_reporters), figsize=(len(min_reporters) * 10, 10))

for j in range(len(min_reporters)):
    for i in range(len(r1)):
        prob = 1 - binom.cdf(min_reporters[j] - 1, d1 * r1[i][0], f1)
        name = 'Reporters: ' + str(r1[i][0]) + ', ' + str(r1[i][1])
        axs.plot(mNeu_real, prob, label=name)
        # axs[j].fill_between(f1, prob(f1, r1[i][0], d1 - d_range1), prob(f1, r1[i][0], d1 + d_range1), alpha=0.2)
    
    axs.plot(mNeu_real, FP, c='k', alpha=0.5, ls='--', label='False Positive')
    axs.fill_between(mNeu_real, FP, y2=0, color='k', alpha=0.5)
    axs.set_xlabel('Fraction (%)', fontsize=16)
    axs.set_ylabel('Probability', fontsize=16)
    axs.set_xscale('log')
    axs.xaxis.set_major_formatter(mtick.FuncFormatter(lambda y, _: '{:g}'.format(y)))
    axs.invert_xaxis()
    axs.set_title('Min Reporters: ' + str(min_reporters[j]), fontsize=18)
    axs.tick_params(axis='x', labelrotation=90)

axs.legend(fontsize=14, loc=(1.03, 0.64))
fig.savefig(out,dpi=300, bbox_inches='tight')
```
