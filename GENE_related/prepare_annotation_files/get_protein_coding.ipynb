```python
import pandas as pd

protein_gene_list='/Users/irffanalahi/Research/Research_update/GENE/processingWithUCSC/mart_export.txt_bed_filterdedWith_protein-coding_gene_name.txt_sorted'
tobefiltered='/Users/irffanalahi/Research/Research_update/GENE/processingWithUCSC/v2/intron_genecode_v36_modified.txt_canonical.txt'

outfile=tobefiltered+"_proteinCodingGene.txt"

protein_gene_listdf=pd.read_csv(protein_gene_list,sep='\t',header=None)
tobefiltereddf=pd.read_csv(tobefiltered,sep='\t')
```
