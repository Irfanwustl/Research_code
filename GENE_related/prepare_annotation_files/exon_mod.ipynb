```python
import pandas as pd
```

```python
file = 'genomic_feature-selected/exon_genecode_v36_modified.txt_canonical.txt_proteinCodingGene.txt'

data = pd.read_csv(file, sep='\t')

df = pd.DataFrame(data)

df
```

```python
df = df.rename(columns = {'transcriptno':'transcriptno_old'})

df
```

```python
new_df = pd.DataFrame(columns = list(df.columns) + ['transcriptno'])

cell_unique = df['hg38.knownCanonical.transcript'].unique()

for i in range(len(cell_unique)):
    df_subset = df[df['hg38.knownCanonical.transcript'] == cell_unique[i]].copy()
    col = []
    for j in range(len(df_subset)):
        if j == 0:
            col.append(0)
        elif (j == len(df_subset) - 1) and (len(df_subset) != 2):
            col.append(2)
        else:
            col.append(1)
            
    df_subset['transcriptno'] = col
    new_df = pd.concat([new_df, df_subset])

new_df
```

```python
new_df.to_csv(file + '_method2.txt', sep='\t', index=False)
```

```python
new_df2 = pd.DataFrame(columns = list(df.columns) + ['transcriptno'])

cell_unique = df['hg38.knownCanonical.transcript'].unique()

for i in range(len(cell_unique)):
    df_subset = df[df['hg38.knownCanonical.transcript'] == cell_unique[i]].copy()
    col = []
    for j in range(len(df_subset)):
        if len(df_subset) < 3:
            col.append(j)
        elif (len(df_subset) % 3 != 0):
            if j < (len(df_subset) // 3):
                col.append(0)
            elif (j >= (len(df_subset) // 3)) and (j <= ((len(df_subset) // 3) * 2)):
                col.append(1)
            elif j > (len(df_subset) // 3) * 2:
                col.append(2)
        else:
            if j < (len(df_subset) // 3):
                col.append(0)
            elif (j >= (len(df_subset) // 3)) and (j < ((len(df_subset) // 3) * 2)):
                col.append(1)
            elif j >= (len(df_subset) // 3) * 2:
                col.append(2)
            
    df_subset['transcriptno'] = col
    new_df2 = pd.concat([new_df2, df_subset])

new_df2
```

```python
new_df2.to_csv(file + '_method1.txt', sep='\t', index=False)
```
