```python
import pandas as pd
infile='/Users/irffanalahi/Research/Research_update/GENE/intron_exon_subset/exon_genecode_v36_modified.txt_canonical.txt_proteinCodingGene.txt'
outfile=infile+"_ZEROtranscript.txt"



indf=pd.read_csv(infile,sep="\t")

indf.head()
```

```python
indf.shape
```

```python
positivezro=indf[(indf['transcriptno']==0) & (indf['direction']=='+')]
positivezro.head()
```

```python
negativegene=indf[indf['direction']=='-']
negativegene.head()
```

```python
negativegenegrouped=negativegene.groupby(['hg38.knownCanonical.transcript'])
negativegeneZerorowlist=[]
for name, group in negativegenegrouped:
    maxindex=[group['transcriptno'].idxmax()]
    negativegeneZerorowlist.append(maxindex)
```

```python
negativegeneZerorowlist=[item for sublist in negativegeneZerorowlist for item in sublist]
```

```python
negativegeneZeroDF=indf.iloc[negativegeneZerorowlist]
negativegeneZeroDF.head()
```

```python
outdf=pd.concat([positivezro,negativegeneZeroDF])
outdf.shape
```

```python
outdf[outdf.duplicated(subset=['#hg38.knownCanonical.chrom_x','hg38.knownCanonical.chromStart_x','hg38.knownCanonical.chromEnd_x','hg38.knownCanonical.transcript'])]
```

```python
outdf.to_csv(outfile,sep="\t",index=False)
```
