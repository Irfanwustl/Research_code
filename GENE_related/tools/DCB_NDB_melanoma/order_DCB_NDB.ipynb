```python
import pandas as pd
import json
infile='/Users/irffanalahi/Research/Research_update/GENE/Someresults/StemGene_for_melanoma_response/melanoma_bg_all_matrix_intesectedwith_stemcell_sig_promoter.txt'
gtfile='/Users/irffanalahi/Research/Research_update/GENE/Someresults/StemGene_for_melanoma_response/prev_and_Em_yalefirstbatch_Response_nohealthy.txt'
indf=pd.read_csv(infile,sep='\t')
indf.head()
```

```python
gtdf=pd.read_csv(gtfile,sep='\t')
gtdf.shape
```

```python
gtdf=gtdf[gtdf['Time point']=='PreTx']
gtdf.shape
```

```python
gtdf['Mixture'] =  gtdf['Mixture'].astype(str)+'.bedgraph_rolled'
gtdf.head()
```

```python
allDCB=gtdf[gtdf['Response']=='DCB']['Mixture'].tolist()
allNDB=gtdf[gtdf['Response']=='NDB']['Mixture'].tolist()
```

```python
corresdict={}
for adcb  in allDCB:
    corresdict[adcb]='DCB'

for andb in allNDB:
    corresdict[andb]='NDB'
    
with open(gtfile+"_dict.txt", 'w') as convert_file:
     convert_file.write(json.dumps(corresdict))
```
