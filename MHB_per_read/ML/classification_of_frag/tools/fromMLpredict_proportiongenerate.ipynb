```python
import pandas as pd
predictfile="/Users/irffanalahi/Research/weekly/for_3_24_21/ML/ML_postprocess/dataprepare/testdata/filnamerename/subtracttrainingdata/predict/onlypredictedfile/mel_dataset_cp2pos_cp1neg.txt_testsize_0.7_model_testfile_allfinal_lenrefCPG_subsetoffinal_cpg1_row_combinedfilename.txt_informative_meltumor_cpg1.txt_notraining.txt_predicted.txt"
celltype='MEL_TUMOR'
outname=predictfile+"_"+celltype+"_proportion.txt"
predictdf=pd.read_csv(predictfile,sep="\t")
predictdf.head()
```

```python
gk =predictdf.groupby('filename')
mixlist=[]
proportionlist=[]
for name, group in gk:
    mixlist.append(name)
    pos=(group[group['mlprediction']==0]).shape[0]
    neg=(group[group['mlprediction']==1]).shape[0]
    
    prop=1.0*pos/(pos+neg)
    
    proportionlist.append(prop)
```

```python
outdf=pd.DataFrame({"Mixture":mixlist,celltype:proportionlist})
outdf.head()
```

```python
outdf.to_csv(outname,sep="\t",index=False)
print("done")
```
