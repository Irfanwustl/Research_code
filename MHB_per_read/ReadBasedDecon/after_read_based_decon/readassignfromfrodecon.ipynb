```python
import pandas as pd
from collections import defaultdict
```

```python
fordeconfile="/Users/irffanalahi/Research/weekly/for_10_8_20/Debug_readcount/resultfrom_rd1/TWCY-1313-1313-BULK-PBMC_sorted_out_fordecon.txt"
outname=fordeconfile+"_readassigned"
fordecondf=pd.read_csv(fordeconfile,sep="\t")

fordecondf['supported_read'] = fordecondf.supported_read.apply(lambda x: x[1:-1].split(','))


fordecondf.head()
```

```python
grouped=fordecondf.groupby('celltype')
```

```python
rawreadcount={}
rawreadcountunique={}
for name, group in grouped:
    print(name)
    allsupportedread=group['supported_read'].tolist()
    
    #print("no empty list.....")
    allsupportedread=[ele for ele in allsupportedread if ele != ['']]
    #print(allsupportedread)
    
    allsupportedread=[item for sublist in allsupportedread for item in sublist]
    
    allsupportedread=[eval(s) for s in allsupportedread]  ###############
    #print(allsupportedread) #####
    
    
    rawreadcount[name]=allsupportedread

    print(len(rawreadcount[name]))
    
    print("unique......")
    allsupportedread=list(set(allsupportedread))
    rawreadcountunique[name]=allsupportedread
    print(len(rawreadcountunique[name]))
    #break

    

with open(outname+"_notunique.txt", 'w') as f:
    print(rawreadcount, file=f)

with open(outname+"_unique.txt", 'w') as f:
    print(rawreadcountunique, file=f)  

    
print("done")
   
```
