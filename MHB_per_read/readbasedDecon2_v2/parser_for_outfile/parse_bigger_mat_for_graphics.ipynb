```python
mastermat="/Users/irffanalahi/Research/weekly/for_10_8_20/graphics/mhb_cd4_cd8/uncollapse_mhb/analyse_readassign_ok/outfolder/alldefault/stat/BULK-PBMC-1389.bg_howsm_single_mode_pp_readALLCpG.txt_mastertable.txt"#/Users/irffanalahi/Research/weekly/for_10_8_20/graphics/forcodedevelopment/data/out/corres_pos_cgr1-1222701_nodup_sorted.bam_rd2out_howsm_single_mode_pp_readALLCpG.txt_mastertable.txt"
outname=mastermat+"_CpGperReadPercelltype"


mastermatdf=pd.read_csv(mastermat,sep="\t",index_col=0)

mastermatdf.head()
```

```python
allreads=mastermatdf.columns
#print(allreads)
```

```python
def nestedlistlength(twodlist):
    alllistlength=[]
    for l in twodlist:
        alllistlength.append(len(l))
    return alllistlength
        
```

```python
outdict={}
for index, row in mastermatdf.iterrows():
    celltype=eval(index)
    #print(celltype)
    readcount=0
    allcpgintheread=[]
    for read in allreads:
        cpglist=eval(row[read])
        if len(cpglist)!=0:
            allcpgintheread.append(cpglist)
            readcount=readcount+1
        
    #print(readcount)
    #print(allcpgintheread)
    Listslength=nestedlistlength(allcpgintheread)
    #print(Listslength)
    
    with open(outname+"_"+celltype+".txt", 'w') as lf:
        for item in Listslength:
            lf.write("%s\n" % item)
    
    allcpgintheread = [x for sublist in allcpgintheread for x in sublist]
    #print(allcpgintheread)
    avglength=(len(allcpgintheread)*1.0)/readcount
    
    
    #print(avglength)
    
    outdict[celltype]=avglength

    
#print(outdict)
with open(outname+"_avg.txt", 'w') as f:
    print(outdict, file=f)
       

print("done")
    
    
```
