```python
import pandas as pd
infile="/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/BL14_all_matrixCin_nr0.4_imputed_rowmean.txt_bg_othermean_backup.txt"

outfile=infile

scorecolumns=['CD4-others','CD8-others','nB-others','NK-others','Mn-others','mNeu-others','m8-others','DC-others','Eo-others','Tr-others','m4-others','Mg-others','Er-others','mB-others']

indf=pd.read_csv(infile,sep="\t")
indf.head()
```

```python
import pybedtools
```

```python
indf[['chrom','start']] = indf.position.str.split(':',expand=True)
indf['start'] = indf['start'].astype(int)
```

```python
indf.head()
```

```python
indf['end']=indf['start']+2
indf.head()
```

```python
indf=indf.drop(['position'],axis=1)
```

```python
indf.head()
```

```python
indf=indf.set_index(['chrom','start','end'])
```

```python
indf.head()
```

```python
indf=indf[scorecolumns]
indf.head()
```

```python
indfbed= pybedtools.BedTool.from_dataframe(indf,header=True,index=True)
```

```python
forinterbed= pybedtools.BedTool('/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/genepromoter_1000_1000_sorted.txt')
```

```python
intersected=indfbed.intersect(forinterbed,header=True,wa=True)
```
