```python
import pandas as pd
from collections import defaultdict
import sys
import time
start_time = time.time()

infile="/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/uisnglinux/BL14_all_matrixCin_nr0.4_imputed_rowmean.txt_bg_othermean_backup_nohead_interwithgeneprom1000_1000.txt"


outfile=infile

indf=pd.read_csv(infile,sep="\t",header=None)



indf.head()
```

```python
allcomuns=['chrom','start','end','CD4','CD8','nB','NK','Mn','mNeu','m8','DC','Eo','Tr','m4','Mg','Er','mB','CD4-others','CD8-others','nB-others','NK-others','Mn-others','mNeu-others','m8-others','DC-others','Eo-others','Tr-others','m4-others','Mg-others','Er-others','mB-others']
```

```python
len(allcomuns)
```

```python
indf.shape
```

```python
indf.columns =allcomuns
indf.set_index(['chrom','start','end'],inplace=True)
indf.head()
```

```python
scorecolumns=['CD4-others','CD8-others','nB-others','NK-others','Mn-others','mNeu-others','m8-others','DC-others','Eo-others','Tr-others','m4-others','Mg-others','Er-others','mB-others']


indf=indf[scorecolumns]
indf.head()
```

```python
negindf=-1*indf
negindf.head()
```

```python
alldelta=[.05,.1,.5,.6,.7,.8,0.9]
```

```python
scorecolumns=negindf.columns
```

```python
celltypeStats=defaultdict(list)

allindex=defaultdict(list)
for score in scorecolumns:

    for delta in alldelta:
        tmpdf=negindf[negindf[score]>=delta]
        allindex[delta]=allindex[delta]+tmpdf.index.tolist()
        
        celltypeStats[score]=celltypeStats[score]+[tmpdf.shape[0]]
        
        
    
```

```python
celltypeStatsdf=pd.DataFrame.from_dict(celltypeStats, orient='index',columns=alldelta)
celltypeStatsdf.index.name = 'Celltype'
celltypeStatsdf.head()
```

```python
celltypeStatsdf.to_csv(outfile+"_stats.txt",sep="\t")
```

```python
alluniqueindex=defaultdict(list)
for key in allindex:
    alluniqueindex[key]=list(set(allindex[key]))
```

```python
selected=0.1
```

```python
sm=indf.loc[alluniqueindex[selected]]
```
