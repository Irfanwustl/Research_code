```python
import pandas as pd
import os

infile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/BL22/moreSM/Thebestref/allctpossepreated/BL22_genepromSM_0.7.txt_TRAININGbestref.txt_withmincol.txt'


outdir=infile+"_ctpos"
os.mkdir(outdir)
indf=pd.read_csv(infile,sep='\t')

indf.head()
```

```python
indf['end']=indf['start']+2
indf.head()
```

```python
grouped=indf.groupby('mincolname')
for name, group in grouped:
    tempdf=group[['chrom','start','end']]
    ctname=name.replace('-others','')
    
    outname=outdir+"/"+ctname
    tempdf.to_csv(outname,sep='\t',index=False, header=False)
    
```
