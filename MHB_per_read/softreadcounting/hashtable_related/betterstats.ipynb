```python
import pandas as pd
import numpy as np
from collections import defaultdict
import sys

infile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/BL22/moreSM/BL22_genepromSM_0.7.txt'
outfile=infile
indf=pd.read_csv(infile,sep="\t",index_col=['chrom','start'])


maxcpg=400
mincpg=100

step=50


indf.head()
```

```python
allctscorename=indf.columns
allctscorename
```

```python
indf['minscoredCT']=(indf[allctscorename]).idxmin(axis=1)
indf.head()
```

```python
indf['minscore']=indf[allctscorename].min(axis=1)
indf.head()
```

```python
totcpglist=np.arange(mincpg, maxcpg+1, step).tolist()
totcpglist
```

```python
sorteddf=indf.sort_values(by=['minscore'])
sorteddf.head()
```

```python
groupedct=sorteddf.groupby('minscoredCT')



strongestdeltaLIST=[]
weakestdeltaLIST=[]


for numcpg in totcpglist:
    tempstats=defaultdict(list)
    tempstatsbestdelta=defaultdict(list)
    tempsmlist=[]
    for name, group in groupedct:
        tempdf=group.head(n=numcpg)
        
        
        
        strongestdelta=-1*(tempdf['minscore'].min())
        
        weekestdelta=-1*(tempdf['minscore'].max())
        
        tempstats[name].append(weekestdelta)
        
        tempstatsbestdelta[name].append(strongestdelta)
        
        tempout=tempdf.drop(['minscoredCT','minscore'],axis=1)
       
        
        tempsmlist.append(tempout)
        
    
    
    
    tempstatsbestdeltaDF=pd.DataFrame.from_dict(tempstatsbestdelta,orient='index',columns=[numcpg])
    
   
    
    tempstatsDF=pd.DataFrame.from_dict(tempstats,orient='index',columns=[numcpg])
    
    
    strongestdeltaLIST.append(tempstatsbestdeltaDF)
    weakestdeltaLIST.append(tempstatsDF)
    
   
        
    
    currentout=pd.concat(tempsmlist)
    
    
    currentout.to_csv(outfile+"_g_"+str(numcpg)+".txt",sep="\t")
    
    
        
        
    
```

```python
strongestdeltaLISTDF=pd.concat(strongestdeltaLIST,axis=1)
weakestdeltaLISTDF=pd.concat(weakestdeltaLIST,axis=1)
```

```python
strongestdeltaLISTDF.to_csv(outfile+"_strongestDELTAg_"+str(mincpg)+"_"+str(maxcpg)+".txt",sep="\t")
weakestdeltaLISTDF.to_csv(outfile+"_weakestDELTAg_"+str(mincpg)+"_"+str(maxcpg)+".txt",sep="\t")
```
