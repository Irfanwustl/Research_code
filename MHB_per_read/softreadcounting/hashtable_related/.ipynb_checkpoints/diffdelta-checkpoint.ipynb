```python
import pandas as pd
infile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/uisnglinux/diffdelta/BL14_all_matrixCin_nr0.4_imputed_rowmean.txt_bg_othermean_backup_nohead_interwithgeneprom1000_1000.txt_sm_0.7.txt'
outfile=infile
indf=pd.read_csv(infile,sep="\t",index_col=['chrom','start'])
#selectedCT=['nB-others','NK-others','mNeu-others','Eo-others','mB-others']
delta_cutoff=-0.8
indf.head()
```

```python
allctscorename=indf.columns
allctscorename
```

```python
selectedCT=allctscorename
```

```python
indf['minscoredCT']=(indf[allctscorename]).idxmin(axis=1)
indf.head()
```

```python
indf['minscore']=indf[allctscorename].min(axis=1)
indf.head()
```

```python
grouped=indf.groupby('minscoredCT')

shoulddeleteindex=[]
for name, group in grouped:
    if name in selectedCT:
        shoulddeleteindex=shoulddeleteindex+group[group['minscore']>delta_cutoff].index.tolist()
        
```

```python
shoulddeleteindexuniq=list(set(shoulddeleteindex))
```

```python
len(shoulddeleteindex)
```

```python
len(shoulddeleteindexuniq)
```

```python
outdf=indf[~(indf.index.isin(shoulddeleteindexuniq))]
```

```python
indf.shape
```

```python
outdf.shape
```

```python
outdf=outdf.drop(['minscoredCT','minscore'],axis=1)
```

```python
outdf.to_csv(outfile+"_diffdelta"+str(delta_cutoff)+".txt",sep="\t")
```
