```python
import pandas as pd
from collections import defaultdict

infile='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/veryefficient/fulltest/allct_nohyper/CD4allrange_NR_1000000_insilmix41_sorted_scored_nohyper.txt'

indf=pd.read_csv(infile,sep="\t",index_col=0)


deltagreaterforpositive=0  ##############>0
allctscorename=indf.columns
indf.head()
```

```python
allctscorename
```

```python
indf['maxscoredCT']=(indf[allctscorename]).idxmax(axis=1)
indf.head()
```

```python
indf['maxscore']=indf[allctscorename].max(axis=1)
indf.head()
```

```python
indf['deltabasedfragassignment']='NotAssigned'
indf.loc[indf['maxscore']>deltagreaterforpositive,'deltabasedfragassignment']=indf.loc[indf['maxscore']>deltagreaterforpositive,'maxscoredCT']
indf.head()
```

```python
set(indf['deltabasedfragassignment'].tolist())
```

```python
ctposscoredict= defaultdict(list)
ctposfragdict= defaultdict(list)
for score in allctscorename:
    deltabasedfragassigned=indf['deltabasedfragassignment'].tolist()
    if score in deltabasedfragassigned:
        temp_posscore=indf.loc[indf['deltabasedfragassignment']==score,'maxscore'].tolist()
        temptotal_posscore=sum(temp_posscore)
        temp_posfrag=len(temp_posscore)
    
    else:
        temptotal_posscore=0
        temp_posfrag=0
        
    ctname=score.replace('-others','')
    
    ctposscoredict[ctname].append(temptotal_posscore)

    ctposfragdict[ctname].append(temp_posfrag)
        
        
```

```python
ctposscoredf=pd.DataFrame.from_dict(ctposscoredict)
ctposfragdf=pd.DataFrame.from_dict(ctposfragdict)
ctposscoredf.head()
```

```python
totalposscore=ctposscoredf.sum(axis=1)[0]

posscore_proportion=ctposscoredf/totalposscore
```

```python
totalposfrag=ctposfragdf.sum(axis=1)[0]

posfrag_proportion=ctposfragdf/totalposfrag
```

```python
ctposscoredf.to_csv(infile+"_posscore.txt",sep="\t",index=False)
ctposfragdf.to_csv(infile+"_posfrag.txt",sep="\t",index=False)

posscore_proportion.to_csv(infile+"_posscore_PROPORTION.txt",sep="\t",index=False)
posfrag_proportion.to_csv(infile+"_posfrag_PROPORTION.txt",sep="\t",index=False)




indf.to_csv(infile+"_binned_stats.txt",sep="\t")
```
