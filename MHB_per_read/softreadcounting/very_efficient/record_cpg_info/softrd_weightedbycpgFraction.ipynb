```python
import pandas as pd
import sys

softRDfile='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/veryefficient/fulltest/CD4allrange_NR_1000000_insilmix60_sorted_scored_nohyper.txt'
cpginfofile='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/veryefficient/fulltest/CD4allrange_NR_1000000_insilmix60_sorted_scored_nohyper_cpgoinfomerged.txt'
score_columns=['CD4-others','CD8-others','nB-others','NK-others','Mn-others','mNeu-others','m8-others','DC-others','Eo-others','Tr-others','m4-others','Mg-others','Er-others','mB-others']
softRDdf=pd.read_csv(softRDfile,sep="\t",index_col=0)
softRDdf.head()
```

```python
cpginfodf=pd.read_csv(cpginfofile,sep="\t",index_col=0)
cpginfodf.head()
```

```python
print(softRDdf.shape)
print(cpginfodf.shape)
```

```python
cpghyperinfodf=cpginfodf[['hyperCpG','LENhyperCpG']]
cpghyperinfodf.head()
```

```python
cpghyperinfodf=cpghyperinfodf[cpghyperinfodf['hyperCpG']!='NotFound']
cpghyperinfodf.head()
```

```python
cpghyperinfodf[cpghyperinfodf.index.duplicated()]
```

```python
mergeddf=softRDdf.merge(cpghyperinfodf,left_index=True,right_index=True,how='left')
if mergeddf.shape[0]!=softRDdf.shape[0]:
    print('merge problem')
    sys.exit(1)
```

```python
mergeddf.head()
```

```python
mergeddf['LENhyperCpG'].fillna(0, inplace=True)
mergeddf.head()
```

```python
mergeddf['total_cpg']=mergeddf['LENhypoCpG']+mergeddf['LENhyperCpG']

mergeddf['LENhypoCpG_BY_total_cpg']=mergeddf['LENhypoCpG']/mergeddf['total_cpg']

max_value = mergeddf['LENhypoCpG_BY_total_cpg'].max()
min_value=mergeddf['LENhypoCpG_BY_total_cpg'].min()
print(max_value)
print(min_value)

if max_value >1 or min_value<0:
    print('max,min error')
    sys.exit(1)

mergeddf.head()
```

```python
for scolumn in score_columns:
    mergeddf[scolumn]=mergeddf[scolumn]*mergeddf['LENhypoCpG_BY_total_cpg']
```

```python

mergeddf.head()
```

```python
outdf=mergeddf[score_columns]
outdf.head()
```

```python
outdf.to_csv(softRDfile+"_cpgweightedforBIN.txt",sep='\t')
mergeddf.to_csv(softRDfile+"_cpgweightedlog.txt",sep='\t')
```
