```python
import pandas as pd


posscorefile='/Users/irffanalahi/Research/Research_update/SoftRD/largerEXP/BL22genepromdelta.7/Realdata/response_prediction/BL22LTMEv2/forrename/BL22LTMEtrainingV2.txt_result_dupindex_binnedstats.pkladjustedScore_minus_avg_CSxOut.txt'
ct_cpgfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/B22withLTME/v2_2steps/generateSM/BL22LTMEtrainingV2.txt_info.txt'

posscoredf=pd.read_csv(posscorefile,sep="\t",index_col='Mixture')

posscoredf.head()
```

```python
ct_cpgdf=pd.read_csv(ct_cpgfile,sep="\t")
ct_cpgdf.head()
```

```python
allct=ct_cpgdf['Celltype'].tolist()
len(allct)
```

```python
for ct in allct:
    posscoredf[ct]=posscoredf[ct]/int(ct_cpgdf.loc[ct_cpgdf['Celltype']==ct,'#CpG'])
```

```python
sumofnormalizedctscore=posscoredf[allct].sum(axis=1)
```
