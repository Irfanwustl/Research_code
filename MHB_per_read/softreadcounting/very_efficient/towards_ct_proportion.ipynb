```python
import pandas as pd


posscorefile='/Users/irffanalahi/Research/Research_update/SoftRD/largerEXP/spike_EXP/cellularfraction/fromserver/allct_towardsSM_dummy_mean_SM.txt_unique_withmincol.txtEPCAM_mean_SM_top1000.txt_result_dupindex_binnedstats.pkl_maxscore_CSxOut.txt'
ct_cpgfile='/Users/irffanalahi/Research/Research_update/SoftRD/largerEXP/spike_EXP/cellularfraction/fromserver/allct_towardsSM_dummy_mean_SM.txt_unique_withmincol.txtEPCAM_mean_SM_top1000.txt_info.txt'

posscoredf=pd.read_csv(posscorefile,sep="\t",index_col='Mixture')

posscoredf.head()
```
