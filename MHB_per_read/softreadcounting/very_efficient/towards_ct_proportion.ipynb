```python
import pandas as pd


posscorefile='/Users/irffanalahi/Research/Research_update/SoftRD/largerEXP/BL22genepromdelta.7/insilico/training/bestrefperfforTEST/ROC_plots/BL22_binned/finalizedforeditorppt/BL22_genepromSM_0.7.txt_TRAININGbestref.txt_result_dupindex_binnedstats.pkladjustedScore_minus_CSxOut.txt'
ct_cpgfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/BL22/moreSM/Thebestref/BL22_genepromSM_0.7.txt_TRAININGbestref.txt_info.txt'

posscoredf=pd.read_csv(posscorefile,sep="\t",index_col='Mixture')

posscoredf.head()
```
