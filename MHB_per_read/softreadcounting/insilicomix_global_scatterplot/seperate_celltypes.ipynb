```python
import pandas as pd 
infile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/BL22/moreSM/Thebestref/heatmaps/deltaheat/linked_nonlinked/BL22_TRAININGbestref_sorted'

outfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/BL22/moreSM/Thebestref/heatmaps/deltaheat/linked_nonlinked/BL22trainingbestrefallct_seperated'

indf=pd.read_csv(infile,sep='\t')

indf.head()
```
