```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]

infile='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/veryefficient/BL14_all_matrixCin_nr0.4_imputed_rowmean.txt_bg_intesectedwith_CD4DMRofBL14atleast.2SM_g1_CD4_3_g2_others_33_withheader.txt_othermean.txt'
metfile='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/veryefficient/BL14_all_matrixCin_nr0.4_imputed_g1_mNeu_3_g2_others_33.txt'


recordfile='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/veryefficient/BL14_atleast.2_top-1_singleCpG_final_record.txt'


celltype='CD4-others'
indf=pd.read_csv(infile,sep="\t",index_col=['chrom','start'])

indf.head()
```

```python
ax = sns.heatmap(indf)
```

```python
indfneg=-1*indf
```

```python
ax = sns.heatmap(indfneg,vmin=-1, vmax=1,cmap='viridis')
```

```python
indfnegSorted=indfneg.sort_values(by=[celltype],ascending=False)
indfnegSorted.head()
```

```python
allcolumns=indfnegSorted.columns
ax =indfnegSorted.plot( y='CD4-others',  label='CD4-others')
for ct in allcolumns:
    if ct=="CD4-others":
        continue
    indfnegSorted.plot( y=ct,  label=ct,ax=ax)
ax.set_ylim([-1,1])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
