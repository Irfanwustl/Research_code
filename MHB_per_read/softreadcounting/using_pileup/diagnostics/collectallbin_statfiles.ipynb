```python
import pandas as pd
import glob
import os

infol='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/softRD_pileup/cd4binnstatsfiles_rebindiffdelta'

suffix='rebinstats.txt'

allstatfiles=glob.glob(infol+"/*"+suffix)




```

```python
filesdf=[]
for fp in allstatfiles:
    tmpdf=pd.read_csv(fp,sep="\t",index_col=0)
    tmpdf['filename']=os.path.basename(fp)
    filesdf.append(tmpdf)
```

```python
df_combined = pd.concat(filesdf, axis=0)
df_combined.head()
```

```python
df_combined.shape
```

```python
nodupindex = df_combined[~df_combined.index.duplicated(keep='first')]
```

```python
nodupindex.shape
```

```python
df_combined.to_csv(infol+"_"+suffix,sep='\t')
nodupindex.to_csv(infol+"_nodupindex_"+suffix,sep='\t')
```
