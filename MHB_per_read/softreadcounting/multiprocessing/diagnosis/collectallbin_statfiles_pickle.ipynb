```python
import pandas as pd
import glob
import os
import time
start_time = time.time()

infol='/Users/irffanalahi/Research/Research_code/gitignorefolder/MHB_per_read/softreadcounting/testtheidea/softRD_pileup/CD4mixture_output_81205_sorted_softMultiprocessing'

suffix='binnedstats.pkl'

allstatfiles=glob.glob(infol+"/*"+suffix)




```

```python
filesdf=[]
for fp in allstatfiles:
    tmpdf=pd.read_pickle(fp)
    tmpdf['filename']=os.path.basename(fp)
    filesdf.append(tmpdf)
```

```python
df_combined = pd.concat(filesdf, axis=0)
df_combined.head()
```

```python
nodupindex = df_combined[~df_combined.index.duplicated(keep='first')]
```
