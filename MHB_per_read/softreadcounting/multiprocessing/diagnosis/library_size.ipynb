```python
import pandas as pd
from collections import defaultdict


infile='/Users/irffanalahi/Research/Research_update/SoftRD/largerEXP/genepromdelta.7/REAL_samples/library_size_corrected/BL14_genepromSM_0.7.txt_bestref.txt_result_dupindex_binnedstats.pkl'

indf=pd.read_pickle(infile)

indf.head()
```

```python
fgrouped=indf.groupby('filename')


libsizedict=defaultdict(list)

for name, group in fgrouped:
    libsizedict[name].append(group.shape[0])
    
```

```python
libsizedict
```

```python
libsizeDF=pd.DataFrame.from_dict(libsizedict, orient='index',columns=['Total_fragment'])
libsizeDF.index.name='Mixture'
libsizeDF
```

```python
libsizeDF.to_csv(infile+"_libsize.txt",sep="\t")
```
