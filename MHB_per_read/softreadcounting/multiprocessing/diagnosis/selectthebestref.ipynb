```python
import pandas as pd
import numpy as np
from collections import defaultdict
import sys

infile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/B22withLTME/towarsbestREF/helpfromBL22/BL22LTMEtraining_genepromSM_0.7.txt'



bestreffile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/B22withLTME/towarsbestREF/helpfromBL22/SM_BL22Trainmixture_ManualLTMETIL13_TOselectbestREF.txt'

bestreffileDF=pd.read_csv(bestreffile,sep="\t",index_col=0)


outfile=infile
indf=pd.read_csv(infile,sep="\t",index_col=['chrom','start'])


indf.head()
```

```python
bestref=bestreffileDF.to_dict()
bestref=bestref['MaxAUCCpG']
bestref
```

```python
allctscorename=indf.columns
```

```python
indf['minscore']=indf[allctscorename].min(axis=1)
indf.head()
```

```python
indf['minscoredCT']=(indf[allctscorename]).idxmin(axis=1)
indf.head()
```

```python
sorteddf=indf.sort_values(by=['minscore'])
sorteddf.head()
```

```python
groupedct=sorteddf.groupby('minscoredCT')


tempsmlist=[]
for name, group in groupedct:
   
    tempdf=group.head(n=bestref[name])
    
    
    tempout=tempdf.drop(['minscoredCT','minscore'],axis=1)
    
    tempsmlist.append(tempout)
    
```

```python
outdf=pd.concat(tempsmlist)

outdf.head()
```

```python
outdf.to_csv(outfile+"_bestref.txt",sep="\t")
```
