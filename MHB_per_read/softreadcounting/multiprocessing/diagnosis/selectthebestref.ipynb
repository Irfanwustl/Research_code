```python
import pandas as pd
import numpy as np
from collections import defaultdict
import sys

infile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/uisnglinux/moresm/BL14_genepromSM_0.7.txt'
outfile=infile
indf=pd.read_csv(infile,sep="\t",index_col=['chrom','start'])


indf.head()
```

```python
bestref={'CD4-others':374,'CD8-others':25,'nB-others':25,'NK-others':30,'Mn-others':50,'mNeu-others':45,'m8-others':25,'DC-others':30,'Eo-others':25,'Tr-others':40,'m4-others':100,'Mg-others':25,'Er-others':100,'mB-others':25}
```

```python
sum(bestref.values())
```

```python
len(bestref)
```

```python
allctscorename=indf.columns
```

```python
indf['minscore']=indf[allctscorename].min(axis=1)
indf.head()
```

```python
indf['minscoredCT']=(indf[allctscorename]).idxmin(axis=1)
indf.head()
```

```python
sorteddf=indf.sort_values(by=['minscore'])
sorteddf.head()
```

```python
groupedct=sorteddf.groupby('minscoredCT')


tempsmlist=[]
for name, group in groupedct:
   
    tempdf=group.head(n=bestref[name])
    
    
    tempout=tempdf.drop(['minscoredCT','minscore'],axis=1)
    
    tempsmlist.append(tempout)
    
```

```python
outdf=pd.concat(tempsmlist)

outdf.head()
```

```python
outdf.shape
```

```python
outdf.to_csv(outfile+"_bestref.txt",sep="\t")
```
