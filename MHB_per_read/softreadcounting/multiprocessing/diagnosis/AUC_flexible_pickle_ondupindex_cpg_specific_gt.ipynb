```python




def calculate_metrics(currentgroup,currentmethods):
   
    
    cell_frame = dict()
    cell_frame_total=dict()
    cellframe_ctotal=dict()
    
    
    
    cell_names = []
    #row_count = 0
    for name, group in currentgroup:
        cell_names.append(name)
        celltype=name.replace('-others','')


        for file in np.unique(inbindf['filename']):
            
            
            df_sub = group[group['filename'] == file]
            y_sub=(df_sub['index'].str.contains(celltype)).astype(int)
            cell_count = np.count_nonzero(y_sub == 1)
          
                
                
            
           # row_count += cell_count
            if file in cell_frame:
                cell_frame[file].append(cell_count / len(y_sub))
                
                
                cell_frame_total[file].append(len(y_sub))
                cellframe_ctotal[file].append(cell_count)
            else:
                cell_frame[file] = []
                cell_frame[file].append(cell_count / len(y_sub))
                
                cell_frame_total[file] = []
                cellframe_ctotal[file] = []
                cell_frame_total[file].append(len(y_sub))
                cellframe_ctotal[file].append(cell_count)





        y=(group['index'].str.contains(celltype)).astype(int) 


        for method in currentmethods:

            pred=(group[method]).to_numpy()


            fpr, tpr, thresholds=metrics.roc_curve(y,pred,pos_label=1)


            tempAUC=metrics.auc(fpr, tpr)

            celltypeAUC[celltype]=celltypeAUC[celltype]+[tempAUC]

            optimal_idx = np.argmax(tpr - fpr)
            optimal_threshold = thresholds[optimal_idx]
            

            
            celltypeoptcutpoint[celltype]=celltypeoptcutpoint[celltype]+[optimal_threshold]
            
            celltypeoptcutpointTPR[celltype]=celltypeoptcutpointTPR[celltype]+[tpr[optimal_idx]]
            celltypeoptcutpointFPR[celltype]=celltypeoptcutpointFPR[celltype]+[fpr[optimal_idx]]
            
            
    
    df_files = pd.DataFrame.from_dict(cell_frame, orient='index',columns=cell_names)
    df_files.index.name = 'Mixture'
    df_files.to_csv(outfile + '_cellcounts_cpgspecific.txt', sep='\t')
    
    cell_frame_totaldf=pd.DataFrame.from_dict(cell_frame_total, orient='index',columns=cell_names)
    cellframe_ctotaldf=pd.DataFrame.from_dict(cellframe_ctotal, orient='index',columns=cell_names)
    
    cell_frame_totaldf.index.name = 'Mixture'
    cellframe_ctotaldf.index.name = 'Mixture'
    
    cell_frame_totaldf.to_csv(outfile + '_cellcounts_cpgspecific_totalfrag.txt', sep='\t')
    cellframe_ctotaldf.to_csv(outfile + '_cellcounts_cpgspecific_ctfrag.txt', sep='\t')
```

```python



celltypeAUC=defaultdict(list)
celltypeoptcutpoint=defaultdict(list)
celltypeoptcutpointTPR=defaultdict(list)
celltypeoptcutpointFPR=defaultdict(list)




allassignedgrouped=allassigned.groupby('deltabasedfragassignment')


methods=['adjustedScore_minus_avg']

calculate_metrics(allassignedgrouped,methods)


```

```python
celltypeAUCdf=pd.DataFrame.from_dict(celltypeAUC, orient='index',columns=methods)
celltypeAUCdf.index.name = 'Celltype'
celltypeAUCdf.head()
```

```python
celltypeoptcutpointdf=pd.DataFrame.from_dict(celltypeoptcutpoint, orient='index',columns=methods)
celltypeoptcutpointdf.index.name = 'Celltype'
celltypeoptcutpointdf.head()
```
