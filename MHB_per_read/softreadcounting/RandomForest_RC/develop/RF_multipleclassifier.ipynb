```python
def common_member(a, b):
    a_set = set(a)
    b_set = set(b)
    if len(a_set.intersection(b_set)) > 0:
        return(True) 
    return(False)
```

```python
grouped=indf.groupby('deltabasedfragassignment')

for name, group in grouped:
    
    celltype=name.replace('-others','')
 
    
  
    notctindex=group.index[group['label']!=celltype].tolist()
    ctindex=group.index[group['label']==celltype].tolist()
    
    if common_member(notctindex,ctindex)==True:
        print("something wrong")
        sys.exit(1)
    

    indf.loc[notctindex,'label2']="NOT"+celltype
    indf.loc[ctindex,'label2']=celltype
    
   
    
    
    
    
```

```python
from sklearn.metrics import RocCurveDisplay

celltypeAUC=defaultdict(list)

alltrainindex=[]
alltestindex=[]

for score in scorecolumns:
    
    tempct=score.replace('-others','')
    
 
  
    indfsingleCT=(indf[indf['deltabasedfragassignment']==score]).copy()
    indfsingleCT.loc[indfsingleCT['label2']==tempct,'label2']=1
    indfsingleCT.loc[indfsingleCT['label2']=='NOT'+tempct,'label2']=0
    

    
    indfsingleCT['label2']=indfsingleCT['label2'].astype(int)
    
   
    
    
    
    Y=indfsingleCT[gtcol]
    X=indfsingleCT[features]
    trainX,testX,trainY,testY =  train_test_split(X,Y,test_size=testratio, random_state=0)
    

    traindf=pd.concat([trainX,trainY],axis=1)
    testdf=pd.concat([testX,testY],axis=1)
    
    traindf.loc[traindf['label2']==1,'label2']=tempct
    traindf.loc[traindf['label2']==0,'label2']='NOT'+tempct
    
    testdf.loc[testdf['label2']==1,'label2']=tempct
    testdf.loc[testdf['label2']==0,'label2']='NOT'+tempct
    
    
    
    
    alltrainindex=alltrainindex+traindf.index.tolist()
    alltestindex=alltestindex+testdf.index.tolist()
    
    
   
    
    RDF_classifier = RandomForestClassifier(oob_score=True,n_estimators=n_trees)
    
    RDF_classifier.fit(trainX, trainY)
    
    
    
    predict_test=RDF_classifier.predict_proba(testX)
    
    
    fpr, tpr, thresholds=metrics.roc_curve(testY,predict_test[:,1],pos_label=1)
    
    


    tempAUC=metrics.auc(fpr, tpr)
    
    
   
    
   


    
    
    celltypeAUC[score]=celltypeAUC[score]+[tempAUC]

    
   
```

```python
alltraindflistDF=indf.loc[alltrainindex]
alltestdflistDF=indf.loc[alltestindex]
```

```python
alltraindflistDF.to_pickle(infile+"_train.pkl")
alltestdflistDF.to_pickle(infile+"_test.pkl")
```

```python
celltypeAUCdf=pd.DataFrame.from_dict(celltypeAUC, orient='index',columns=['AUC'])
celltypeAUCdf.index.name = 'Celltype'
celltypeAUCdf.to_csv(infile+"_RFAUC.txt",sep="\t")
```
