```python
import pandas as pd

oursequence='/Users/irffanalahi/Research/Research_update/troubleshootoffbyoneissue/promdatareadyfolder_thatCD8TIL_.6.2_andothercpgs/promdataready_all_matrix_intwith_BL22LTMEtraining_NewSMTIL85_posprob_pad2.txt'

othersequene='/Users/irffanalahi/Research/Research_update/troubleshootoffbyoneissue/BL22promfolder_thatCD8TIL_.6.2_andothercpgs/BL22_all_matrix_promdata_intwith_BL22LTMEtraining_NewSMTIL85_posprob_pad2.txt'


ourseqdf=pd.read_csv(oursequence,sep='\t')
ourseqdf.head()
```

```python
otherseqdf=pd.read_csv(othersequene,sep='\t')
otherseqdf.head()
```

```python
lessthanone=[]
greaterthanone=[]
forwhichthisishappenning=[]
for index, row in ourseqdf.iterrows():
    currentchrom=row['chrom']
    currentstart=row['start']
    templessthanone=otherseqdf[(otherseqdf['chrom']==currentchrom) & (otherseqdf['start']==currentstart-1)]
    tempgreaterthanone=otherseqdf[(otherseqdf['chrom']==currentchrom) & (otherseqdf['start']==currentstart+1)]
    
    lessthanone.append(templessthanone)
    greaterthanone.append(tempgreaterthanone)
    
    if templessthanone.shape[0]>0 or tempgreaterthanone.shape[0]>0:
        forwhichthisishappenning.append(row)
    
    
```

```python
lessthanoneDF=pd.concat(lessthanone)
print(lessthanoneDF.shape)
lessthanoneDF.head()
```

```python
greaterthanoneDF=pd.concat(greaterthanone)
print(greaterthanoneDF.shape)
greaterthanoneDF.head()
```

```python
lessthanoneDF.to_csv(othersequene+"_lessthanone.txt",sep='\t',index=False)
greaterthanoneDF.to_csv(othersequene+"_greaterthanone.txt",sep='\t',index=False)
```

```python
forwhichthisishappenningDF=pd.DataFrame(forwhichthisishappenning)
print(forwhichthisishappenningDF.shape)
forwhichthisishappenningDF.head()
```

```python
forwhichthisishappenningDF.to_csv(oursequence+"_whereithappening.txt",sep='\t',index=False)
```
