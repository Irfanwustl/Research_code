```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.special import expit
def sigmoid_shape(x, high, low):
    return expit(x) *(high-low)+low
```

```python
myx = np.arange(-8., 8., 0.2)
sig = sigmoid_shape(myx,1,0)
```

```python
plt.plot(myx,sig)
plt.show()
```

```python
sig = sigmoid_shape(myx,5,0)
plt.plot(myx,sig)
plt.show()
```

```python
sig = sigmoid_shape(myx,5,-5)
plt.plot(myx,sig)
plt.show()
```
