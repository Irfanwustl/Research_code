```python
####tested for hypo only####
import pandas as pd
import numpy as np
from collections import defaultdict




infile = #'/Users/irffanalahi/Research/Research_update/SM/melcfdref/bestworkingref/genomic_analysis/pbltilmel_mincp2outcutoff_addedcol_genomic_feature_withrepeat_header_celltypeseperated/otheranalysis/promdataready_all_matrixCin_nr0.5_imputed_g1_TIL_9_g2_others_9.txt_q0.05_diff0.1_hypo.txt_PBL_TIL_MEL_TUMOR.txt'

percentiles=[99,95,75]

diffcol='diff'
qcol='q'

outname=infile+"_percentile"

indf=pd.read_csv(infile,sep='\t')
indf.head()
```

```python
absdiffcol='abs_'+diffcol
indf[absdiffcol]=indf[diffcol].abs()
neglogqcol='neglog'+qcol
indf[neglogqcol]=-np.log10(indf[qcol])
```

```python
indf.head()
```

```python
#np.percentile(a,perc)
outdict=defaultdict(list)
perdflist=[]
for perc in percentiles:
    tempdiffcutoff=np.percentile(indf[absdiffcol],perc)
    tempqcutoff=np.percentile(indf[neglogqcol],perc)
    

    
    tempdf=indf[(indf[absdiffcol]>=tempdiffcutoff) & (indf[neglogqcol]>= tempqcutoff)].copy()
    tempdf['percentile']=perc
    outdict['perc'].append(perc)
    outdict['#DMR'].append(tempdf.shape[0])
    outdict['#CpG'].append(tempdf['#cpg'].sum())
    
    perdflist.append(tempdf)

    
outperfdf=pd.concat(perdflist)
outdf=pd.DataFrame.from_dict(outdict)

outdf.to_csv(outname+"_stat.txt",sep="\t",index=False)
```
