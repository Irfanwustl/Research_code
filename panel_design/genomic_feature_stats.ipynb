```python
import pandas as pd
from collections import defaultdict
infile="/Users/irffanalahi/Research/Research_update/SM/melcfdref/bestworkingref/genomic_analysis/pbltilmel_mincp2outcutoff_addedcol_genomic_feature_withrepeat_header_celltypeseperated/otheranalysis/genomic_region_stats/promdataready_all_matrixCin_nr0.5_imputed_g1_TIL_9_g2_others_9.txt_q0.05_diff0.1_hypo.txt_PBL_TIL_MEL_TUMOR.txt_percentile_percedf.txt_merged.txt"

indf=pd.read_csv(infile,sep="\t")

indf.head()
```

```python
unique_perce=list(set(indf['percentile'].tolist()))
unique_regions=list(set(indf['Region'].tolist()))

print(unique_perce)
print(unique_regions)
```

```python
outdict=defaultdict(list)
for perce in unique_perce:
    for region in unique_regions:
        tempdf=indf[(indf['percentile']==perce) & (indf['Region']==region)]
        outdict['percentile'].append(perce)
        outdict['region'].append(region)
        outdict['#DMR'].append(tempdf.shape[0])
        outdict["#cpg"].append(tempdf['#cpg'].sum())
        

outdf=pd.DataFrame.from_dict(outdict)
outdf.to_csv(infile+"_genomestats.txt",sep="\t",index=False)
```
