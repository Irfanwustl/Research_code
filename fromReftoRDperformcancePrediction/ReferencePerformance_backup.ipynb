```python
DMRdfct=DMRdf[allct]
DMRdfct.head()
```

```python
DMRdfct=1-DMRdfct
DMRdfct.head()
```

```python
print(DMRdfct.shape)
print(mixturegtdf.shape)
mixturegtdfT=mixturegtdf.transpose(copy=True)
mixturegtdfT=mixturegtdfT.rename(index={celltype+"_real":celltype})
mixturegtdfT.head()
```
