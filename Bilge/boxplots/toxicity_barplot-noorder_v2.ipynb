```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import wilcoxon, mannwhitneyu
import matplotlib 
plt.rc('font', family='Arial')
```

```python
cutoff = 2
#order = ['Memory B cell',  'CD8 effector','NK','Naive B cell','Naive CD8','CD8 central memory','Plasma','Mono', 'CD8 effector memory', 'Tregs','Naive CD4',
        #'CD4 effector memory','CD4 central memory']
file = '/Users/irffanalahi/Research/Research_update/SoftRD/largerEXP/BL22genepromdelta.7/Realdata/irAE/benchmark_withcelfie/boxplots/celfie-on-melanoma_CSxOut.txt_irAEgt_PreTX.txt_ctrename.txt'
data = pd.read_csv(file, sep='\t', index_col=0)
df = pd.DataFrame(data)

#order = ['Memory B',  'Naive B','NK','CD8 TCM','Mac (M2)','Mac (M1)','Mac (M0)','PC','PMN','Eos','Mono', 'CD8 TEMRA', 'CD8 TEM', 'Tregs','Naive CD8 T','Naive CD4 T',
 #       'CD4 TEM']
order=df.columns.tolist()
order.remove('Toxicity')
```

```python
conditions = [(df['Toxicity'] <= cutoff), (df['Toxicity'] > cutoff)]
values = ['toxicity <= ' + str(cutoff), 'toxicity > ' + str(cutoff)]
df['Toxic'] = np.select(conditions, values)
df
```

```python
p_vals = {}

REALpvals_irf={}
for cell in order:
    p_val_list = []
    if cell not in ['Toxicity', 'Toxic']:
        multi = 1
        for i, toxicity in enumerate(values):
            y = np.array(df[df['Toxic'] == toxicity][cell])
            p_val_list.append(y)
        if np.median(p_val_list[0]) >= np.median(p_val_list[1]):
            multi = -1
        p_vals[cell] = -np.log10(mannwhitneyu(p_val_list[0], p_val_list[1], alternative='two-sided')[1]) * multi
        REALpvals_irf[cell]=(mannwhitneyu(p_val_list[0], p_val_list[1], alternative='two-sided')[1])*multi
p_vals
```

```python
type(p_vals)
p_vals={k: v for k, v in sorted(p_vals.items(), key=lambda item: item[1])}
print(p_vals)
order=p_vals.keys()
print(order)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 10))
hfont = {'fontname':'Helvetica'}
#matplotlib.rc('xtick', labelsize=20) 
#matplotlib.rc('ytick', labelsize=60) 
for cell in order:
    y = p_vals[cell]
    if y <= 0:
        ax.barh(cell, y, color='lightgrey')
    else:
       # if cell=='CD4 TEM':
        #    ax.barh(cell, y, color='r')
        #else:
        ax.barh(cell, y, color='lightgrey')

# for label in ax.get_xticklabels():
#     label.set_fontproperties('Helvetica')
ax.tick_params(axis="y", labelsize=16)
ax.tick_params(axis="x", labelsize=13)
ax.axvline(linewidth=1, color='k')
ax.axvline(x=-np.log10(0.05), linewidth=1.5, color='k', ls='--')
ax.annotate('P < 0.05', (1.4, 8), fontsize=16,rotation=90)
#ax.annotate('P < 0.05', (1.4, 8), fontsize=16,rotation=90, **hfont)
ax.set_xlabel('-log10(P)',fontsize=20)
#ax.set_xlabel('-log10(P)',fontsize=20, **hfont)
ax.set_xlim([-1.75,1.75])
#ax.annotate('Severe\niRAE', xy=(0.7, -0.062), xycoords='axes fraction', xytext=(1.13, -0.089), fontsize=18, ha='center',
#            arrowprops=dict(arrowstyle="<-", color='r', lw=2), **hfont)

#ax.annotate('No Severe\niRAE', xy=(0.3, -0.062), xycoords='axes fraction', xytext=(-0.19, -0.089), fontsize=18, ha='center',
#            arrowprops=dict(arrowstyle="<-", color='b', lw=2), **hfont)

ax.annotate('Severe\nirAE', xy=(0.7, -0.062), xycoords='axes fraction', xytext=(1.13, -0.089), fontsize=18, ha='center',
            arrowprops=dict(arrowstyle="<-", color='r', lw=2))

ax.annotate('No severe\nirAE', xy=(0.3, -0.062), xycoords='axes fraction', xytext=(-0.19, -0.089), fontsize=18, ha='center',
            arrowprops=dict(arrowstyle="<-", color='b', lw=2))
#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)
#ax.spines['bottom'].set_visible(False)
#ax.spines['left'].set_visible(False)
plt.savefig(file+'_cutoff_'+str(cutoff)+'_response_barplot.pdf', bbox_inches='tight')
```
