```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

phenofile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/Majorlineage/Genomic_feature/any_relation_using_genomic_feature_and_noiserate/MajorLineage_PHENOCLASS.txt'


#####Actually file to test#####
crcfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/Majorlineage/Genomic_feature/any_relation_using_genomic_feature_and_noiserate/BLUprom_all_matrix_folder_int_CRCpbl_rankedv1_inflectionpos_sortedby_CRCpbl_rankedv1_inflectionpos/BLUprom_all_matrix_intwith_Mono_CpGdelta_info_faster.txt_5000_forheatunderlyingdata_ranked.txt_pos.txt_sorted.txt'


######Actually SM files#####
blu_genefile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/Majorlineage/Genomic_feature/CRC_inflection/Mono_CpGdelta_info_faster.txt_5000_forheatunderlyingdata_ranked.txt_pos.txt__row_combined.txt'

blu_smfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/Majorlineage/Genomic_feature/any_relation_using_genomic_feature_and_noiserate/CRC_PBLprom_all_matrix_folder_int_CRCpbl_rankedv1_inflectionpos_sortedby_CRCpbl_rankedv1_inflectionpos/CRC_PBLprom_all_matrix_intwith_Mono_CpGdelta_info_faster.txt_5000_forheatunderlyingdata_ranked.txt_pos.txt_sorted.txt'


pheno=pd.read_csv(phenofile,sep='\t',index_col=0,header=None)

crcdf=pd.read_csv(crcfile,sep='\t',index_col=['chrom','start','end'])
celltype='Mono'

crcdf.head()
```

```python
pheno.head()
```

```python
def rowmean(imputedfile):
    rowmeanlist = []

    for i in range(pheno.shape[0]):
        classes = pheno.iloc[i, :]
        class1 = (classes == 1).tolist()
        class1_mean = imputedfile.iloc[:, class1].mean(axis=1)

        temdf =class1_mean.to_frame(pheno.index[i])



        rowmeanlist.append(temdf)


    rowmeandf=pd.concat(rowmeanlist,axis=1)  #########if necessary can save in future


    print("row mean done")

    return rowmeandf
```

```python
crcrowmean=rowmean(crcdf)
crcrowmean.head()
```

```python

blu_genedf=pd.read_csv(blu_genefile,sep='\t',index_col=['chrom','start','end'])
blu_genedf.head()
```

```python
crcrowmean_mergeewith_blu_genedf=crcrowmean.merge(blu_genedf,left_index=True,right_index=True,how='inner')
crcrowmean_mergeewith_blu_genedf.head()
```

```python
crcrowmean_mergeewith_blu_genedf.drop(['3','4','5'],axis=1,inplace=True)
```

```python
crcrowmean_mergeewith_blu_genedf.drop_duplicates(inplace=True)
```

```python
crcrowmean_mergeewith_blu_genedf.head()
```

```python
crcrowmean_mergeewith_blu_genedf.sort_values(by=['Region'],inplace=True)
```

```python
crcrowmean_mergeewith_blu_genedf.head()
```

```python
ax = sns.boxplot(y=celltype, x="Region", data=crcrowmean_mergeewith_blu_genedf)
sns.stripplot(y=celltype, x="Region", data=crcrowmean_mergeewith_blu_genedf)
ax.set_title(celltype)
ax.set_ylabel('Methylation')
ax.set_ylim([0,1])
plt.savefig(crcfile+'_methylation.pdf')
```

```python
blu_smdf=pd.read_csv(blu_smfile,sep='\t',index_col=['chrom','start','end'])
blu_sm_mean=rowmean(blu_smdf)
blu_sm_mean.head()
```

```python
allmerge=crcrowmean_mergeewith_blu_genedf.merge(blu_sm_mean,left_index=True,right_index=True,how='inner')
allmerge.drop_duplicates(inplace=True)
allmerge.head()
```

```python
allmerge[celltype+"_x"+'-'+celltype+"_y"]=allmerge[celltype+"_x"]-allmerge[celltype+"_y"]
allmerge.sort_values(by=['Region'],inplace=True)
```

```python
ax = sns.boxplot(y=celltype+"_x"+'-'+celltype+"_y", x="Region", data=allmerge)
sns.stripplot(y=celltype+"_x"+'-'+celltype+"_y", x="Region", data=allmerge)
ax.set_title(celltype)
ax.set_ylabel('data-SM')

plt.savefig(crcfile+'_difference.pdf')
```
