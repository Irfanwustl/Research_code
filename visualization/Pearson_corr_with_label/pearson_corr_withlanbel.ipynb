```python
from scipy.stats import pearsonr
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename="/Users/irffanalahi/Research/Research_code/visualization/manage_gt/makeallinonefile/data/py_50_merged_gt3_manualonefile.txt"

df=pd.read_csv(filename,sep="\t")

x='ground_truth'
y='predicted'
```

```python
df.corr()
```

```python
corr = pearsonr(df[x], df[y])
#corr = [np.round(c, 5) for c in corr]
print(corr)
text='R=%s \nP=%s' % (corr[0], corr[1])
```

```python

g=sns.scatterplot(df[x], df[y],hue=df['celltype'],edgecolor='none')
g.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), ncol=1)
ax=sns.regplot(df[x], df[y],scatter=False)
ax.text(0,0.6, text, fontsize=12)
savename=filename+"_"+x+"_"+y+".pdf"
plt.savefig(savename,dpi=300,bbox_inches="tight")
plt.show()
#sns.jointplot(data=df, x=x, y=y, kind='reg', hue=df['celltype'])
```
