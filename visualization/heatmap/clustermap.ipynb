```python
import pandas as pd
import seaborn as sns; sns.set()
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
file="/Users/irffanalahi/Research/weekly/for_10_1_20/SM/downsample_cpg_tileup/Allpblsubset/allpblSmdata_bg_allmat_cyberin_header"
outref=file+"_ref.txt"
df=pd.read_csv(file,sep="\t",index_col=0)
trans=False
df.head()
```

```python
plt.figure(figsize=(12,10))

mycmap = plt.cm.viridis

mycmap.set_bad("dimgrey")

cg = sns.clustermap(df,cmap=mycmap,method='centroid')
#cg=sns.clustermap(df, cmap="viridis")
cg.ax_row_dendrogram.set_visible(False)
cg.ax_col_dendrogram.set_visible(False)
```

```python
outdf=df.iloc[cg.dendrogram_row.reordered_ind,:]
outdf.to_csv(outref,sep="\t")
```
