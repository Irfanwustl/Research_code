```python
import pandas as pd
import re
from collections import defaultdict

infile="data/blwithneu_hypo_stat.txt_0.01_2_g50.txtWithSurrounding.txt_1cpgnc.txt_result_nocov_mincpg5_ABSreadcount_divisioned.txt_CSxOut.txt"

outfile=infile+"_gt.txt"
##mixname:whatWeWant
cellmapping={'mono':'CD14_real','nB':'CD19_real','cd4naive':'CD4_real','cd8naive':'CD8_real','NK':'CD56_real','mNeu':'Neu_real'}

querstr='insilmix_NR_\d+_'


indf=pd.read_csv(infile,sep="\t")
indf.head()
```

```python

firstmixname=indf['Mixture'][0]

numberofcells=firstmixname.count("_s_")


for i in range(numberofcells):
    querstr=querstr+"(.+?)_s_(.+?)_"



    
```

```python
rowdict=defaultdict(list)
for index, row in indf.iterrows():
    
    mixname=row['Mixture']
    m = re.search(querstr, mixname)
  
    rowdict['Mixture'].append(mixname)
    j=1
    while j<= numberofcells*2:
        ct=m.group(j)
        j=j+1
        prop=m.group(j)
        rowdict[ct].append(prop)
        j=j+1
    
rowdict=dict(rowdict)
```

```python
outdf=pd.DataFrame.from_dict(rowdict)
outdf.head()
```

```python
outdf.rename(cellmapping,inplace=True,  errors='raise',axis=1)
outdf.head()
```

```python
outdf.to_csv(outfile,sep="\t",index=False)
```
