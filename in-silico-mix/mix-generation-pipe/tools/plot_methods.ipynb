```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
og = '/Users/irffanalahi/Research/Research_update/in-silico/mNeu.txt'
new_met = '/Users/irffanalahi/Research/Research_update/in-silico/mixture_output_newmethod.txt'
prev_met = '/Users/irffanalahi/Research/Research_update/in-silico/mixture_output_previousmethod.txt'

outname='/Users/irffanalahi/Research/Research_update/in-silico/mNeu.txt'+".pdf"

data_og = pd.read_csv(og, sep='\t')
data_new = pd.read_csv(new_met, sep='\t', index_col=0)
data_prev = pd.read_csv(prev_met, sep='\t', index_col=0)

df_og = pd.DataFrame(data_og)
df_new = pd.DataFrame(data_new)
df_prev = pd.DataFrame(data_prev)

display(df_og)
display(df_new)
display(df_prev)
```

```python
np.array(df_og.index)
```

```python
x = np.array(df_og.index)

fig, axs = plt.subplots(1, len(df_og.columns), figsize=(10*len(df_og.columns), 10))

for i in range(len(df_og.columns)):
    cell = list(df_og.columns)[i]
    y_og = np.array(df_og[cell])
    y_new = np.array(df_new[cell])
    y_prev = np.array(df_prev[cell])
    axs[i].plot(x, y_og, label='Targeted proportion', ls='--', lw=3)
    axs[i].plot(x, y_new, label='New method', ls='-.', lw=3)
    axs[i].plot(x, y_prev, label='Previous method', ls=':', lw=3)
    axs[i].set_title(cell)
    axs[i].set_yscale('log') ######

plt.legend(loc=(1.02, 0.87), fontsize=14) 
fig.savefig(outname,dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

for i in range(len(df_og.columns)):
    cell = list(df_og.columns)[i]
    y_og = np.array(df_og[cell])
    y_new = np.array(df_new[cell])
    y_prev = np.array(df_prev[cell])
    y_comp = np.zeros(len(y_og))
    for i in range(len(y_og)):
        ls = np.array([y_og[i], y_new[i], y_prev[i]])
        y_comp[i] = np.median(ls)
    ax.plot(x, y_comp, label=cell)
    
plt.legend(loc=(1.03, 0.45), fontsize=14)
```
