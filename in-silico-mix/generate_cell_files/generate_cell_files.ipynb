```python
import numpy as np
import pandas as pd
import random as ran
import os
```

```python
file_cn = 'Data/BL14_PhenoClass.txt'

transpose = True
header = None
index_col= True

'''
lower_lim = 0.00001
upper_lim = 0.01
'''

lower_lim = -5
upper_lim = -2

repetition = 5

mix_num = 10

outfol = file_cn+'_proportions'
```

```python
if transpose:
    data_cn = pd.read_csv(file_cn, sep='\t', index_col=0, header=header)
else:
    data_cn = pd.read_csv(file_cn, sep='\t', header=header)

df_cn = pd.DataFrame(data_cn)

if transpose:
    df_cn = df_cn.T

df_cn.columns
```

```python
def generate_list(cell_name, cell_list, val):
    length = len(cell_list)
    vals = np.zeros(length)
    r = [ran.random() for i in range(1,length)]
    s = sum(r)
    r = [ (i/s) * (100 - val) for i in r ]
    name_seen = False
    for i in range(length):
        if cell_list[i] != cell_name:
            if name_seen == False:
                vals[i] = r[i]
            elif name_seen:
                vals[i] = r[i - 1]
        elif cell_list[i] == cell_name:
            vals[i] = val
            name_seen = True
    return vals
```

```python
#mixtures = np.linspace(lower_lim, upper_lim, num=mix_num) * 100
mixtures = np.logspace(lower_lim, upper_lim, num=mix_num) * 100
mixtures
```

```python
os.mkdir(outfol)
df_final = pd.DataFrame(columns = list(df_cn.columns))
```

```python
# os.mkdir(outfol)

for cell in list(df_cn.columns):
    df_out = pd.DataFrame(columns = list(df_cn.columns))
    for val in mixtures:
        for i in range(repetition):
            arr = generate_list(cell, list(df_cn.columns), val)
            while 0.0 in arr:
                arr = generate_list(cell, list(df_cn.columns), val)
            df_out.loc[len(df_out.index)] = arr
    display(df_out)
    df_out=df_out.round(6)
    df_final = pd.concat([df_final, df_out])
    
    
    df_out.to_csv(outfol + '/' + cell + '.txt', sep='\t', index=False)
```
