```python
import numpy as np
import pandas as pd
import random as ran
import os
```

```python
file_cn = 'Data/BL14_PhenoClass_forNoNutest.txt'

transpose = True
header = None
index_col= 0

mix_num = 10

outfile = file_cn+'_cell_files_random.txt'
```

```python
data_cn = pd.read_csv(file_cn, sep='\t', index_col=index_col, header=header)

df_cn = pd.DataFrame(data_cn)

if transpose:
    df_cn = df_cn.T

df_cn.columns
```

```python
def generate_list(cell_list):
    length = len(cell_list)
    vals = np.zeros(length)
    r = [ran.random() for i in range(length)]
    s = sum(r)
    r = [ (i/s) * 100 for i in r ]
    return r
```

```python
# os.mkdir(outfol)

# for cell in list(df_cn.columns):
df_out = pd.DataFrame(columns = list(df_cn.columns))
for j in range(mix_num):
    arr = generate_list(list(df_cn.columns))
    while 0.0 in arr:
        arr = generate_list(list(df_cn.columns))
    print(np.sum(arr))
    df_out.loc[len(df_out.index)] = arr
display(df_out)
df_out.to_csv(outfile, sep='\t', index=False)
```
