```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
f = np.logspace(-1, -6, num=1000)
r =[100,216,47,126,68,16,23,107,34,93,110,340,31,29] #[3, 6, 17, 34, 67]
d = 30 #5000
d_range = 5 #1000
```

```python
def prob(f, r, d):
    prod = d * r
    dec = (1 - f) ** (prod)
    return 1 - 0.95 * dec
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
for i in range(len(r)):
    plt.plot(f, prob(f, r[i], d), label='r = ' + str(r[i]))
    plt.fill_between(f, prob(f, r[i], d - d_range), prob(f, r[i], d + d_range), alpha=0.2)

plt.legend(fontsize=14) #, loc='lower left')
plt.xlabel('f', fontsize=16)
plt.ylabel('Probability', fontsize=16)
ax.set_xscale('log')
ax.invert_xaxis()
```
