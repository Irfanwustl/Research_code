```python
import pandas as pd
SMfile="/Users/irffanalahi/Research/Research_update/in-silico/background_pattern/readnumber_1000000/mincpg3/BL14_atleast.2_top-1_singleCpG_final_assignedref_uniq.txt"

outfile=SMfile+"_noFPcpgs.txt"

FPcpgsfile="/Users/irffanalahi/Research/Research_update/in-silico/background_pattern/readnumber_1000000/mincpg3/somefinalfiles_uniqueFPcpgs.txt"

FPcpgsdf=pd.read_csv(FPcpgsfile,sep="\t",header=None)

SMdf=pd.read_csv(SMfile,sep="\t")
SMdf.head()
```

```python
FPcpgsdf.head()
```

```python
FPcpgsdf[['chrom','start']] = FPcpgsdf[0].str.split(pat=":",expand=True)
FPcpgsdf['start']=FPcpgsdf['start'].astype(int)
FPcpgsdf.drop(columns=[0],axis=1,inplace=True, errors='raise')
FPcpgsdf.head()
```

```python
outdf=pd.merge(SMdf, FPcpgsdf, on=['chrom','start'], how="outer", indicator=True)
outdf.head()
```

```python
outdf = outdf.loc[outdf["_merge"] == "left_only"].drop("_merge", axis=1)
outdf.head()
```

```python
print(SMdf.shape)
print(outdf.shape)
```

```python
outdf.to_csv(outfile,sep="\t",index=False)
```
