```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import fsolve
```

```python
adj = 0.95
out = 'probability.pdf'
```

```python
def prob(f, r, d):
    prod = d * r
    dec = (1 - f) ** (prod)
    return 1 - adj * dec

def reg(x, f, d, p):
    return 1 - adj * ((1 - f) ** (x * d)) - p

def depth(x, f, r, p):
    return 1 - adj * ((1 - f) ** (x * r)) - p

def f_val(x, d, r, p):
    return 1 - adj * ((1 - x) ** (d * r)) - p
```

```python
fig, axs = plt.subplots(1, 4, figsize=(40, 10))

f = np.logspace(-1, -3, num=1000)  ###############################
r = [3,30, 60, 170, 340, 670]
d = 30
d_range = 0
for i in range(len(r)):
    axs[0].plot(f, prob(f, r[i], d), label='r = ' + str(r[i]))
    axs[0].fill_between(f, prob(f, r[i], d - d_range), prob(f, r[i], d + d_range), alpha=0.2)

axs[0].legend(fontsize=14, loc='lower left')
axs[0].set(xlabel='f', ylabel='p')
axs[0].set_xscale('log')
axs[0].invert_xaxis()

p = np.linspace(0.9, 0.99, 5)
d_range = 10
for i in range(len(p)):
    y_ax = np.zeros(len(f))
    y_min = np.zeros(len(f))
    y_max = np.zeros(len(f))
    for j in range(len(f)):
        sol = fsolve(reg, 0.5, args=(f[j], d, p[i]))
        sol_min = fsolve(reg, 0.5, args=(f[j], d - d_range, p[i]))
        sol_max = fsolve(reg, 0.5, args=(f[j], d + d_range, p[i]))
        y_ax[j] = sol[0]
        y_min[j] = sol_min[0]
        y_max[j] = sol_max[0]
    
    axs[1].plot(f, y_ax, label='p = ' + str(p[i]))
    axs[1].fill_between(f, y_min, y_max, alpha=0.2)

axs[1].legend(fontsize=14, loc='upper left')
axs[1].set(xlabel='f', ylabel='r')
axs[1].set_xscale('log')
axs[1].invert_xaxis()

r = 30
r_range = 5
for i in range(len(p)):
    y_ax = np.zeros(len(f))
    y_min = np.zeros(len(f))
    y_max = np.zeros(len(f))
    for j in range(len(f)):
        sol = fsolve(depth, 100, args=(f[j], r, p[i]))
        sol_min = fsolve(depth, 100, args=(f[j], r - r_range, p[i]))
        sol_max = fsolve(depth, 100, args=(f[j], r + r_range, p[i]))
        y_ax[j] = sol[0]
        y_min[j] = sol_min[0]
        y_max[j] = sol_max[0]
    
    axs[2].plot(f, y_ax, label='p = ' + str(p[i]))
    axs[2].fill_between(f, y_min, y_max, alpha=0.2)

axs[2].legend(fontsize=14, loc='upper left')
axs[2].set(xlabel='f', ylabel='d')
axs[2].set_xscale('log')
axs[2].invert_xaxis()

p = np.linspace(0.8, 0.99, num=1000)
r = [30, 60, 170, 340, 670]
d = 30
d_range = 10
for i in range(len(r)):
    y_ax = np.zeros(len(p))
    y_min = np.zeros(len(p))
    y_max = np.zeros(len(p))
    for j in range(len(p)):
        sol = fsolve(f_val, 0.000001, args=(d, r[i], p[j]))
        sol_min = fsolve(f_val, 0.000001, args=(d - d_range, r[i], p[j]))
        sol_max = fsolve(f_val, 0.000001, args=(d + d_range, r[i], p[j]))
        y_ax[j] = sol[0]
        y_min[j] = sol_min[0]
        y_max[j] = sol_max[0]
    
    axs[3].plot(p, y_ax, label='r = ' + str(r[i]))
    axs[3].fill_between(p, y_min, y_max, alpha=0.2)

axs[3].legend(fontsize=14, loc='upper right')
axs[3].set(xlabel='p', ylabel='f')
axs[3].set_yscale('log')
axs[3].invert_xaxis()

fig.savefig(out,dpi=300, bbox_inches='tight')
```
