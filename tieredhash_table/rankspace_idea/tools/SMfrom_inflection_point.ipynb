```python
import pandas as pd
largeSMfile = '/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/rankspace_idea/inflection_point/Blige/ranckedCTv2_fol/rankedCTv2_SM_top1000_SM_unique'

infofile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/rankspace_idea/inflection_point/Blige/ranckedCTv2_fol/rankedCTv2_index.txt'

largesmdf=pd.read_csv(largeSMfile,sep='\t',index_col=['chrom','start'])
largesmdf.head()
```

```python
largesmdf['mincolname']=largesmdf.idxmin(axis=1)
largesmdf.head()
```

```python
infodf=pd.read_csv(infofile,sep='\t')
#infodf['Unnamed: 0']
infodf.head()
```

```python
grouped=largesmdf.groupby('mincolname')
outlist=[]
for name, group in grouped:
   
    tempct=name.replace('-others','')
    corresnumber=(infodf[infodf['Unnamed: 0']==tempct]['Index']).tolist()[0]
   
    tempdf=group.head(n=corresnumber+1)
    
    outlist.append(tempdf)
```

```python
outdf=pd.concat(outlist)
outdf=outdf.drop(['mincolname'],axis=1)
```

```python
outdf.to_csv(infofile+"_inflectionSM.txt",sep='\t')
```
