```python
import pandas as pd 
import sys
import os.path
import shutil

import re

import glob
import os
```

```python
def listdir_nohidden(path):
    return glob.glob(os.path.join(path, '*'))
```

```python
infol='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/BL22_groupv2/em4_cm4/OneSet/em4_int_em4_thresholdpos_allthresholdcombinations_oneset'
linecountfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/BL22_groupv2/em4_cm4/OneSet/em4_int_em4_thresholdpos_allthresholdcombinations_oneset_Shoonhsin_avg_and_linecount.txt'


avgvalcutoff=0.6
lincountcutoff=100

outfol=infol+'_'+str(avgvalcutoff)+'_'+str(lincountcutoff)

os.makedirs(outfol)

allfiles=listdir_nohidden(infol)

allfiles[0]
```

```python
lincounttdf=pd.read_csv(linecountfile,sep='\t')
lincounttdf.head()
```

```python
lincounttdfcutoff=lincounttdf[(lincounttdf['Average Value']>=avgvalcutoff) & (lincounttdf['Line Count']>=lincountcutoff)]
lincounttdfcutoff.shape
```

```python
for file in allfiles:
    if os.path.basename(file) in lincounttdfcutoff['Filename'].tolist():
        
        outfile=outfol+'/'+os.path.basename(file)
        shutil.copyfile(file, outfile)
```
