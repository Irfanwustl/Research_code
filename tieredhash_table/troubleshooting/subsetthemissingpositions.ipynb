```python
import pandas as pd
bigfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/CD8TIL_reproduce/comparewiththat.6.2/that.6_.2_fixed.txt'
smallfile='/Users/irffanalahi/Research/Research_update/SoftRD/hashtable_stats/tieredApproach/CD8TIL_reproduce/comparewiththat.6.2/g1_CD8_TIL_g2_CD8_melPBLCD8_TIL_-0.6_int_g1_CD8_TIL_g2_TcellCD8_TIL_-0.2_int_g1_CD8_TIL_g2_MelTumorCD8_TIL_-0.2_int_g1_CD8_TIL_g2_BcellCD8_TIL_-0.2_int_g1_CD8_TIL_g2_MyloidCD8_TIL_-0.2_int_g1_CD8_TIL_g2_PMNlikeCD8_TIL_-0.2'


bigdf=pd.read_csv(bigfile,sep='\t')
bigdf.head()
```

```python
smalldf=pd.read_csv(smallfile,sep='\t',header=None)
smalldf.rename(columns={0:'chrom',1:'start',2:'end'},inplace=True)
#smalldf=smalldf[['chrom','start']]
smalldf.head()
```

```python
print(bigdf.shape)
print(smalldf.shape)
```

```python
int_df = bigdf.merge(smalldf, how ='inner', on =['chrom','start','end'])
int_df.shape
```

```python
int_df.to_csv(bigfile+"_common.txt",sep='\t',na_rep='NA',index=False)
```

```python
missing_df = pd.concat([bigdf,int_df]).drop_duplicates(keep=False)
missing_df.shape
```
