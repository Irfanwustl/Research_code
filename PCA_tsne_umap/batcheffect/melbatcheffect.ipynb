```python
import pandas as pd
import sys

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import ast
from sklearn.manifold import TSNE
import umap
import matplotlib as mpl
infile="/Users/irffanalahi/Research/Research_update/Dataqc/batcheffect/melhealthy/tomergwithgt_mergewithmelanoma_noah_Response.txt/melhealthycfDNAbgrolled_all_matrix_nona_cin_rand10000_transposed.txt"
indf=pd.read_csv(infile,sep="\t",index_col=0)
whichdecomp="PCA"
```

```python
############################# PCA ############################
def pca_comp(stdData):
	pca = PCA()
	principalComponents = pca.fit_transform(stdData)
	return principalComponents



####################tsne###############
def tsne_comp(stdData):
	tsne = TSNE(random_state=100)
	principalComponents = tsne.fit_transform(stdData)
	return principalComponents


####################umap###############
def umap_comp(stdData):
	um = umap.UMAP(random_state=100)
	principalComponents = um.fit_transform(stdData)
	return principalComponents
```

```python
indf.head()
```

```python
X=indf.drop(['batch','Time point','Response'],axis=1)
X.head()
```

```python
standardizedData=StandardScaler().fit_transform(X)
```

```python
################# call decomposition ###########
if whichdecomp=="PCA":
	projected=pca_comp(standardizedData)
elif whichdecomp=="tSNE":
    projected=tsne_comp(standardizedData)
elif whichdecomp=="UMAP":
    projected=umap_comp(standardizedData)
else:
    print ("wrong reduction")
```

```python
############### draw ######################

import seaborn as sns; sns.set()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
mpl.rcParams['figure.dpi']= 300



g=sns.scatterplot(x=projected[:, 0], y=projected[:, 1],hue=indf['batch'], edgecolor='none',style=indf['Response'])

g.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), ncol=1)
plt.title(whichdecomp)




plt.xlabel('C1')
plt.ylabel('C2')

#plt.ylim(-10,10)
#plt.xlim(-10,0)

savename_samplename=infile+"_"+whichdecomp+".pdf"
plt.savefig(savename_samplename,dpi=300,bbox_inches="tight")
plt.show()
```
