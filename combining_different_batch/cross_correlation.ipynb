```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
infile='/Users/irffanalahi/Research/Research_update/SoftRD/largerEXP/OurCD8TIL_ranked/melanoma_EMseq_secondlane/EM_seconbatch_result/rawscore_analysis/CD8TIL_pbl_tum_all_matrix_allout_ranked_SM_top100_SM_unique_result_dupindex_binnedstats.pkl_NOpat_cg1_CSxOut.txt'

indf=pd.read_csv(infile,sep='\t',index_col='Mixture')
indf.head()
```

```python
droppedindf=indf.drop(['Toxicity','Response','YU ID'],errors='ignore',axis=1)
droppedindf.head()
```

```python
corr = droppedindf.corr(method='pearson')

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,cmap='viridis',vmin=-.5,vmax=1)
plt.savefig(infile + '_CT_crosscorrelation.pdf', bbox_inches='tight')
```
