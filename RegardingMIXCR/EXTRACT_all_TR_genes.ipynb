```python
import pandas as pd
UCSC_genetrack='/Users/irffanalahi/Research/Research_update/GENE/TCR_VDJ/All_Genecode_V40_gene.txt'

UCSC_genedf=pd.read_csv(UCSC_genetrack,sep='\t')

UCSC_genedf.head()
```

```python
TRgenes=UCSC_genedf[UCSC_genedf['name2'].str.startswith('TR')].copy()
TRgenes.head()
```

```python
TRgenes['padded_txStart']=TRgenes['txStart']-1000
TRgenes['padded_txEnd']=TRgenes['txEnd']+1000
```

```python
chrs = ["chr" + str(i) for i in range(1, 23)]
chrs.append("chrX")
chrs.append("chrY")
chrs
```

```python
TRgenes=TRgenes[TRgenes['chrom'].isin(chrs)]
```
