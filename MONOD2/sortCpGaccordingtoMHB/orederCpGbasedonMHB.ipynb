```python
import pandas as pd
```

```python
cpgMHBfile="/Users/irffanalahi/Research/weekly/for_9_10_20/mhb_wgbs_dmr/cpg2/default/uncollapse/heatmap/bgrolled_all_matrix_intesectedwith_smblmhbcpg2_g50_hypo_mhb.txt_ordered_dropped.txt_ciberin_ourdata.txt_inverted"
mhbreffile="/Users/irffanalahi/Research/weekly/for_9_10_20/mhb_wgbs_dmr/cpg2/default/uncollapse/heatmap/smdata_bg_allmat_cyberin_header_ref_myfix.txt_mhb"
outname=cpgMHBfile+"_ordered"

refdf=pd.read_csv(mhbreffile,header=None,sep="\t")
cpgdf=pd.read_csv(cpgMHBfile,sep="\t")
```

```python
refdf=refdf.rename(columns={0: "MHBchr",1:"MHBstart",2:"MHBend" })
refdf.head()
```

```python
cpgdf.head()
```

```python
def ordercpg(dfmhb,dfcpg,outfile):
    allorderedrows=[]
    for index, row in dfmhb.iterrows():
        for indexcpg, rowcpg in dfcpg.iterrows():
            if row["MHBchr"]==rowcpg["MHBchr"] and row["MHBstart"]==rowcpg["MHBstart"] and row["MHBend"]==rowcpg["MHBend"]:
               
                allorderedrows.append(rowcpg)
                
                
    outdf=pd.DataFrame(allorderedrows)
    outdf.to_csv(outfile,index=False,sep="\t",na_rep='NA')
    outdfdropped=outdf.drop(["MHBchr","MHBstart","MHBend"],axis=1)
    outdfdropped.to_csv(outfile+"_dropped",index=False,sep="\t",na_rep='NA')
```

```python
ordercpg(refdf,cpgdf,outname)
print("done")
```
