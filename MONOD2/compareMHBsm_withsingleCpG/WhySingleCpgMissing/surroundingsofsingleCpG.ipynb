```python
smbgposition="/Users/irffanalahi/Research/weekly/for_9_10_20/comparewithSingleCpG/whySingleCpGmissing/blleukog50_hypo.txt_bg"
correspondingallmat="/Users/irffanalahi/Research/weekly/for_9_10_20/comparewithSingleCpG/whySingleCpGmissing/wgbsblwithneu_all_matrix"
radius=65
outname=smbgposition+"_with_"+os.path.basename(correspondingallmat)+"_radius_"+str(radius)+".txt"

smdf=pd.read_csv(smbgposition,sep="\t", header=None)

smdf=smdf.rename(columns={0: "chrom",1:"start",2:"end" })
corresdf=pd.read_csv(correspondingallmat,sep="\t")
```

```python
def calculatesurrounding(sdf,alldf,distance,outfile):
    alldictlist=[]
    for index, row in sdf.iterrows():
            upward=row["start"]-distance
            loward=row["start"]+distance
            currentchr=row["chrom"]
            
           
            selecteddf=alldf[(alldf["chrom"]==currentchr ) & (alldf["start"]>=upward) & (alldf["end"]<=loward)]
            if selecteddf.shape[0]==0:
                print("errorr")
            currentdict={"chrom":row["chrom"],"start":row["start"],"end":row["end"],"surrondingCpG":selecteddf.shape[0]-1}
            alldictlist.append(currentdict)
            
    outdf=pd.DataFrame(alldictlist)
    
    outdf.to_csv(outfile,index=False,sep="\t")
```

```python
calculatesurrounding(smdf,corresdf,radius,outname)
print("done")
```
