```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
infilename="/Users/irffanalahi/Research/weekly/for_9_3_20/test/a_na.txt_Ciberin2"
df=pd.read_csv(infilename,sep="\t",index_col=0)

df.head()
```

```python
def cor_plot(indf,meth,inname):
    cor=indf.corr(method = meth)
    cor.to_csv(inname+"_"+meth+".txt",sep="\t")
    plt.figure(figsize=(35,35))
    ax=sns.heatmap(cor, annot=True,cmap='viridis')
```

```python
cor_plot(df,"pearson",infilename)
```

```python
cor_plot(df,"spearman",infilename)
```
