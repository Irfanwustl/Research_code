```python
import pandas as pd
import glob, os

infolder='/Users/irffanalahi/Research/Research_update/for_backgroundcalculation/try1/distribution_of_Hyper/allsmdistribution/sm.9'
out=infolder
```

```python
files = glob.glob(infolder+'/*.txt')
filesdf=[]
for fp in files:
    tmpdf=pd.read_csv(fp,sep="\t",index_col=['chrom','start','end'])
    tmpdf['filename']=os.path.basename(fp)
    filesdf.append(tmpdf)
```

```python
df_combined = pd.concat(filesdf, axis=0)
df_combined.head()
```

```python
df_combined_onerow=df_combined.drop(['Minimum Value','EPCAM-others','FakeCell-others','filename'],axis=1)
df_combined_onerow.head()
```

```python
df_combined_onerow.shape
```

```python
outdf=df_combined_onerow.melt()
outdf.shape
```

```python
outdf.head()
```

```python
outdf.to_csv(out+"_stacked.txt",sep='\t')
```
