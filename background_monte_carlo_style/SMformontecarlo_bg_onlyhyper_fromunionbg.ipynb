```python
import pandas as pd
infile='/Users/irffanalahi/Research/Research_update/for_backgroundcalculation/try1/distribution_of_Hyper/BL22EPCAM_prom_all_matrix_NO_NA_head.txt_withmincol.txt'

outfile=infile

howmanySM=3
celltype='EPCAM-others'

totalcpg=5
avgdelta=-0.5


indf=pd.read_csv(infile,sep='\t')

indf.head()
```

```python
indf.rename(columns={0: 'chrom', 1: 'start',2:'end',3:'value'},inplace=True)

indf[celltype]=avgdelta
indf['FakeCell-others']=0.9
indf.head()
```

```python
indf=indf[indf['value']>=abs(avgdelta)]
```

```python
for i in range(howmanySM):
    
    #tempsm=indfrequiredcols.sample(n=totalcpg)
    tempsmfull=indf.sample(n=totalcpg)
    tempsm=tempsmfull[['chrom','start',celltype,'FakeCell-others']]
    tempsm.to_csv(outfile+"_randomSM_"+str(i)+".txt",sep='\t',index=False)
    tempsmfull.to_csv(outfile+"_randomSM_"+str(i)+"_corressful.txt",sep='\t',index=False)
    
```
